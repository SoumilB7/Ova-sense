# CNN Lifecycle Visualization

```python
import cv2
import numpy as np
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
bat_sz = 32
pix_sz = 500
epochs = 30
l_r = 0.0001
device = torch.device('cpu')
```

```python
# CNN 
class CancerDetectionCNN(nn.Module):
    def __init__(self):
        super(CancerDetectionCNN, self).__init__()
        self.conv_layers = nn.Sequential(
            
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32), # in order to send in normalized activations
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),

            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256 * (pix_sz // 16) * (pix_sz // 16), 512), 
            nn.ReLU(),
            nn.Dropout(0.5),  
            nn.Linear(512, 1),  
            nn.Sigmoid() 
        )
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x


```

```python
menopause = 1           # 0 fro pre menopause and 1 for post menopause
```

```python
# Load trained model
if menopause:
    model = CancerDetectionCNN()
    model.load_state_dict(torch.load('model_archives/CustomChip/cancer_detection_model_post.pth'))
    model.eval()
else:
    model = CancerDetectionCNN()
    model.load_state_dict(torch.load('model_archives/CustomChip/cancer_detection_model_pre.pth'))
    model.eval()
```

```python

# Load and preprocess the input image
img_path = 'image_v2.png'
transform = transforms.Compose([
    transforms.Resize((500, 500)),
    transforms.ToTensor()
])

image = Image.open(img_path).convert('RGB')
input_image = transform(image).unsqueeze(0)
# image
plt.figure(figsize=(3, 3))
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
img_path = "image_v2.png"
pil_image = Image.open(img_path).convert("RGB")

def detect_faulty_sample(size=500):
    pil_image_c = pil_image.resize((size, size))
    image = cv2.cvtColor(np.array(pil_image_c), cv2.COLOR_RGB2BGR)
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    center = (size//2, size//2)
    radius = size//2 - 40

    def sector_mask(angle_start, angle_end):
        mask = np.zeros((size, size), dtype=np.uint8)
        cv2.ellipse(mask, center, (radius, radius),
                    0, angle_start, angle_end, 255, -1)
        cv2.circle(mask, center, 30, 0, -1)
        return mask

    mask_positive = sector_mask(180, 240)
    mask_negative = sector_mask(240, 300)

    activity_positive = cv2.mean(hsv[:,:,2], mask=mask_positive)[0]
    activity_negative = cv2.mean(hsv[:,:,2], mask=mask_negative)[0]

    faulty = False
    if activity_positive < 20:
        faulty = True
    if activity_negative > 40:
        faulty = True

    return faulty, activity_positive, activity_negative


faulty, pos_act, neg_act = detect_faulty_sample()

if faulty:
    print(f" Faulty sample detected! (Positive={pos_act:.1f}, Negative={neg_act:.1f})")
else:
    print(f" Sample looks good. (Positive={pos_act:.1f}, Negative={neg_act:.1f})")

plt.figure(figsize=(3, 3))
plt.imshow(pil_image)
plt.title('Original Image')
plt.axis('off')
plt.show()

```

```python
activation = {}

def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach()
    return hook

model.conv_layers[0].register_forward_hook(get_activation('conv1'))
model.conv_layers[4].register_forward_hook(get_activation('conv2'))
model.conv_layers[8].register_forward_hook(get_activation('conv3'))
model.conv_layers[12].register_forward_hook(get_activation('conv4'))

with torch.no_grad():
    output = model(input_image)
```

```python
if output.item()<0.5:
    print(f"The predicted condition is : No cancer")
else:
    print(f"The predicted condition is : Cancer")
```

```python
for name, act in activation.items():
    num_feature_maps = act.shape[1]
    fig, axes = plt.subplots(1, min(num_feature_maps, 8), figsize=(16, 4))
    fig.suptitle(f'Feature Maps after layer {name}')

    for i in range(min(num_feature_maps, 8)):
        axes[i].imshow(act[0, i].cpu(), cmap='viridis')
        axes[i].axis('off')

    plt.show()
```
